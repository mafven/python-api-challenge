```python
# Dependencies
import requests
from api_keys import weather_api_key
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import citipy
import scipy
from datetime import date
```

#### Randomly selected + 500 unique (non-repeat) cities based on latitude and longitude.

```python
#Get latitude and longitudes for 2000 ramdom cities (note - Latitude 0 to 90 north and south / longitude 0 to 180 east and west )
m_cities_long_lat_df = pd.DataFrame({"longitudes" : (np.random.uniform(-1.8,1.8,2000)*100), "latitudes" :(np.random.uniform(-0.9,0.9,2000)*100)})
m_cities_long_lat_df.count()
```

```python
from citipy import citipy
cities_list = []
latitudes_list = []
longitudes_list = []

for i in range(0,len(m_cities_long_lat_df['latitudes'])):
    latitudes = m_cities_long_lat_df.iloc[i]['latitudes']
    longitudes = m_cities_long_lat_df.iloc[i]['longitudes']
    city = citipy.nearest_city(latitudes,longitudes)
    name = city.city_name
    print(f'{city} + {name}')
    cities_list.append(name)
    latitudes_list.append(latitudes)
    longitudes_list.append(longitudes)
```

```python
cities_coor_df = pd.DataFrame({"city" : cities_list ,
                              "latitudes" :latitudes_list,
                            "longitudes" : longitudes_list})

cities_coor_df 

#drop any duplicates cities
final_cities_list= cities_coor_df.drop_duplicates("city", keep="first")
final_cities_list
cities = list(final_cities_list["city"])
final_cities_list.count()
```

#### Perform a weather check on each of the cities using a series of successive API calls.

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="
```

```python
# Loop through the list of cities and perform a request for data on each

cities = list(final_cities_list["city"])
city_name = []
latitudes = []
longitudes =[]
temperature_f =[]
humidity = []
cloudiness = []
wind_speed_mph = []
country = []
date = []

for city in cities:
        response = requests.get(query_url + city).json()
        try:
            city_name.append(response["name"])
        except:
            print("name not available") 
            
        try:
            latitudes.append(response["coord"]["lat"])
        except:
            print("latitude not available")
            
        try:
            longitudes.append(response["coord"]["lon"])
        except:
            print("longitude not available")           
        
        try:
            temperature_f.append(response["main"]["temp_max"])
        except:
            print(" temperature not available")
        
        try:
            humidity.append(response["main"]["humidity"])
        except:
            print("humidity  not available")
        
        try:
            cloudiness.append(response["clouds"]["all"])
        except:
            print(" cloud data not available")
        
        try:
            wind_speed_mph.append(response["wind"]["speed"])
        except:
            print("wind speed data not available")

        try:
            country.append(response["sys"]["country"])
        except:
            print("country not available")
            
        try:
            date.append(response["dt"])
        except:
            print("date not available")
```

```python
#create a data frame from cities, lat, Log, temp, hum, cloud,wind

weather_data= {"City" : city_name,
               "Latitudes":latitudes,"Longitudes":longitudes,
               "Temperature" : temperature_f,"Humidity" : humidity,
               "Cloudiness" :cloudiness,"Wind_Speed":wind_speed_mph,"Country": country,"Date" : date}

weather_data_df = pd.DataFrame(weather_data)
weather_data_df.count()
weather_data_df.head()
```

```python
# Save the DataFrame as a csv
weather_data_df.to_csv("final_cities3.csv", encoding="utf-8", index=False)
```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
northern_hemisphere
x_values_temperature_N = northern_hemisphere["Latitudes"]
y_values_temperature_N = northern_hemisphere["Temperature"]


from scipy import stats
# # Perform a linear regression on temperature vs. latitude
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values_temperature_N, y_values_temperature_N)

# Get regression values
regress_values_temp_N = x_values_temperature_N * slope + intercept
# print(regress_values_temp)

# Create line equation string
line_eq_temp_N = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(line_eq_temp_N)
```

```python
plt.scatter(x_values_temperature_N, y_values_temperature_N,color = "steelblue", edgecolors = "royalblue",alpha=0.8)
plt.plot(x_values_temperature_N,regress_values_temp_N,"r-")

# Label plot and annotate the line equation
plt.title('Northern Hemisphere',fontsize=13)
plt.xlabel('Latitude')
plt.ylabel('Temperature')
plt.annotate(line_eq_temp_N,(1,2),fontsize=15,color="red")

# Print r value
print(f"The r-value is: {rvalue**2}")

plt.savefig("NH_temp_latitude.png")
# Show plot
plt.show()
```
